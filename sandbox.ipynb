```python
import pandas as pd
```

```python
df = pd.read_csv('results.csv')

column_dict = dict(zip([x for x in df.columns], ['index', 'timestamp', 'identifier', 'token', 'ori_seg', 'des_seg']))
df = df.rename(columns=column_dict)
df.sample(5)
df = df.sort_values('index')
```

```python
def identifier_filter(df, identifier):
    return df[df['identifier'].str.contains(identifier)]

def assert_size(df1, df2):
    try:
        df1_size = df1.count()[0] 
        df2_size = df2.count()[1]
        assert(df1_size == df2_size)
    except:
        raise Exception('{0} != {1}'.format(df1_size, df2_size))
        
df1 = identifier_filter(df, './corpora/wiki/en/processed_20200319155305799110')
df2 = pd.read_csv('./corpora/wiki/en/processed/test.txt', sep='\t', header=None)
assert_size(df1,df2)
```
